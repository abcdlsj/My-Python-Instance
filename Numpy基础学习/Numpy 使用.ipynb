argsort 用于将数组中的元素排序,返回数组中的索引值

````python
a=np.array([[2,4,6,11],
           [5,8,6,2]])
````

````python
a.argsort()
````

````python
a.argsort(axis=0)
````

````python
a.argsort(axis=1)
````

理解:

````python
x=np.array([2,3,4,5,-19,11])
````

````python
y=np.argsort(x)
````

````python
y
````

这里我们可以看到argsort()函数是将x中的元素从小到大排列，提取其对应的index(索引号)

比如:最小的是x[4]=-19,所以y[0]=4

同理:最大的是x[5]=11,所以y[5]=5

下面是官方案例:
```python
 One dimensional array:一维数组

>>> x = np.array([3, 1, 2])
>>> np.argsort(x)
array([1, 2, 0])

Two-dimensional array:二维数组

>>> x = np.array([[0, 3], [2, 2]])
>>> x
array([[0, 3],
[2, 2]])

>>> np.argsort(x, axis=0) #按列排序
array([[0, 1],
[1, 0]])

>>> np.argsort(x, axis=1) #按行排序
array([[0, 1],
[0, 1]])

>>> x = np.array([3, 1, 2])
>>> np.argsort(x) #按升序排列
array([1, 2, 0])
>>> np.argsort(-x) #按降序排列
array([0, 2, 1])
>>> x[np.argsort(x)] #通过索引值排序后的数组
array([1, 2, 3])
>>> x[np.argsort(-x)]
array([3, 2, 1])
另一种方式实现按降序排序：

>>> a = x[np.argsort(x)]
>>> a
array([1, 2, 3])
>>> a[::-1]
array([3, 2, 1])
```

# 广播

````python
a=np.array([[1,2,3],
            [5,6,8],
            [2,3,4],
            [33,42,1]])
````

````python
b=np.array([1,2,3])
````

````python
for i in range(4):
    a[i,:]+=b
````

````python
##广播就很简单了:
a+b
````

参考网上文章[通俗易懂的解释numpy中的广播](https://blog.csdn.net/xiang_freedom/article/details/77968164)

对两个数组，分别比较他们的每一个维度（若其中一个数组没有当前维度则忽略），满足：

- 当前维度的值相等
- 当前维度的值有一个是1

若条件不满足，抛出“ValueError: frames are not aligned”异常。 
输出数组的维度是每一个维度的最大值，广播将值为1的维度进行“复制”、“拉伸”，如图所示
